## Step 2 — Save to CSV 💾
`expenses` ko `expenses.csv` mein save karo using `csv.DictWriter`.
Fields: `item`, `category`, `amount`.

````python
expenses = [
            {"item": "apple", "category": "fruite", "amount": 120 },
            {"item": "banana", "category": "fruite", "amount": 90},
            {"item": "pine apple", "category": "fruits", "amount": 250},
            {"item": "grapes", "category": "fruits", "amount": 150}
]

print("expenses:", expenses)
print(expenses )
````

````python
import csv
# TODO: define fields
fields = ["item", "category", "amount"]

# TODO: write expenses.csv (writeheader + loop writerow)
with open("expenses.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields, extrasaction="ignore")
    writer.writeheader()
    for e in expenses:
        writer.writerow(e)

print("💾 Saved to expenses.csv")
````

## Step 4 — Total spending 💰
Loop chala kar saare amounts ka **total** nikaalo aur print karo.

> Tip: ek `total = 0` banao, phir loop mein `total = total + ...` (ya `total += ...`).
> `DictReader` se aane wali value text hoti hai — `float(row["amount"])` use karo.

````python
# TODO: loop over the CSV rows and add up amounts
total = 0

with open("expenses.csv", "r") as f:
    reader = csv.DictReader(f)
# for row in reader:
#     total = total + float(row["amount"])
    for row in reader:
        total = total + float(row["amount"])
print("Total spent: Rs", total)
````

## Step 5 — Unique categories 🏷️
Saari categories ek list mein nikaalo (loop se), phir `set()` se **unique** categories print karo.
Aur `len()` se batao kitni alag categories thीं.

````python
# TODO: collect categories into a list using a loop, then set() them
categories = []

# TODO: print unique categories and how man
with open("expenses.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        categories.append(row["category"])

unique_categories = set(categories)
print("Unique categories:", unique_categories)
print("How many:", len(unique_categories))
````
