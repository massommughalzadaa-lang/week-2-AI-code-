## Exercise 4 — Write a CSV 📝
`books` list ko `books.csv` mein save karo using `csv.DictWriter`.
Steps: import csv → fields banao → open in "w" → writeheader → loop + writerow.

````python
import csv

books = [
   {'title': 'Python 101', 'author': 'Ali', 'year': 2026},
    {'title': 'Data Magic', 'author': 'Nabiha', 'year': 2025},
    {'title': 'AI Basics', 'author': 'Omar', 'year': 2024}
]

# TODO: Define fields (the dictionary keys)
fields = ['title', 'author', 'year']

with open("books.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    for book in books:
        writer.writerow(book)

print("books.csv save ho gayi ✅")

````

## Exercise 5 — Read the CSV back 📖
`books.csv` ko `csv.DictReader` se padho aur har row print karo.

````python
# TODO: read books.csv and print each row's title + author
with open("books.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(row['title'], "by", row['author'])
````

## Exercise 6 — Mini practice for the project
User se `input()` lekar **2 students** ka data lo (name + marks),
list of dicts mein daalo, phir CSV `students.csv` mein save karo.

*(Fixed 2 entries — straight line, no loop for input. No if/else.)*

````python
import csv

# User se 2 students ka data lo
name1 = input("Enter student 1 name: ")
marks1 = input("Enter student 1 marks: ")

name2 = input("Enter student 2 name: ")
marks2 = input("Enter student 2 marks: ")

# List of dicts
students = [
    {'name': name1, 'marks': marks1},
    {'name': name2, 'marks': marks2}
]

# Save to students.csv
with open("students.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=['name', 'marks'])
    writer.writeheader()
    writer.writerow(students[0])
    writer.writerow(students[1])

print("students.csv save ho gayi ✅")
````
