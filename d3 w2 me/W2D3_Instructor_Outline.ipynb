## 1. Memory vs Disk 🧠💾
- Ab tak ka data **memory (RAM)** mein tha → program band, data khatam.
- **File** disk par save hoti hai → permanent. Dobara khol sakte hain.

Aaj hum data ko file mein likhna aur file se padhna seekhenge.

## 2. Writing & reading text files ✍️
`with open(...)` — safe tareeqa. File apne aap band ho jaati hai.
- `"w"` = write (purana mita kar naya likho)
- `"r"` = read
- `"a"` = append (end mein add karo)

````python
with open("MyFirstFile.txt", "w") as f:
    f.write("Hi, THis is Asim!\n")
    f.write("THis is second line of code!\n")
````

````python
with open("MyFirstFile.txt", "r") as f:
    content=f.read()
print(content)
````

````python
# Writing to a text file
with open("hello.txt", "w") as f:
    f.write("Assalam-o-Alaikum!\n")
    f.write("This is my first file.\n")

print("File ban gayi ✅")
````

````python
# Reading the whole file back
with open("hello.txt", "r") as f:
    print(f.read())

# print(content)
````

````python
with open("hello.txt", "r") as s:
    content = s.read()

print("Before strip function")
print(content)
print("After strip function")
print(content.strip())
````

````python
# Reading line by line with a loop ⭐
with open("hello.txt", "r") as f:
    for i in f:
        print(i.strip())   # .strip() removes extra newline

````

````python
with open("MyFirstFile.txt", "a") as f:
    f.write("THis is my 3rd line of code\n")
````

````python
Bio = "     THe boy is going to karachi,     and he is going karachi    "
print(Bio.strip())
````

## 3. What is CSV? 📊
**CSV = Comma-Separated Values.** Ek simple text file jahan:
- Har **line = ek row**
- Values **comma se separate** hoti hain
- Pehli line aksar **headers** hoti hai (column names)

```
name,phone,city
Ali,0342-3987710,Karachi
Nabiha,0300-2222222,Lahore
```

> Yeh wohi format hai jo **Excel** kholta hai aur jo **Pandas** (Week 5) load karta hai. Bohat important! 🎯

## 4. Writing CSV — `csv.DictWriter` 📝
Humara data **list of dictionaries** hai (kal wala!). `DictWriter` usay seedha CSV bana deta hai.

````python
import csv

contacts = [
    {"name": "Ali",      "phone": "0342-3987710", "city": "Karachi"},
    {"name": "Nabiha",   "phone": "0300-2222222", "city": "Lahore"},
    {"name": "Fatimah",  "phone": "0300-3333333", "city": "Quetta"},
]

# Column names (dictionary ki keys)
headers = ["name", "phone", "city"]

````

````python
print(contacts)
print(type(contacts))
````

````python
print(contacts[0])
print(contacts[1])
print(contacts[2])
print(contacts[0]["name"])
print(contacts[0]["phone"])
````

````python
print(headers)
print(headers[0])
print(headers[1])
````

````python
contacts
````

````python
with open("contacts.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=headers)
    # writer.writeheader()          # pehli line: headers
    for c in contacts:
        writer.writerow(c)        # har dict = ek row

print("contacts.csv save ho gayi ✅")
````

> 💡 `newline=""` zaroori hai CSV likhte waqt — warna Windows par khaali lines aa jaati hain.

## 5. Reading CSV back — `csv.DictReader` 📖

````python
import csv

with open("contacts.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        # har row ek dictionary banti hai — keys = headers
        print(row["name"], "|", row["phone"], "|", row["city"])
````

## 6. 🏗️ BUILD — Contact Book → CSV (in-class project)
Aaj hum jo banayenge:
- User se kuch contacts `input()` se lenge (fixed number — straight line, no loop-for-input)
- Unhe list of dictionaries mein rakhenge
- CSV file mein save karenge
- Wapas padh kar dikhayenge

> Note: hum entries ko ginti ke hisaab se lenge (e.g. 3 contacts) — koi "add another?" wala if/while nahi (woh Week 3).

````python
import csv

print("📒 Add 3 contacts to your Contact Book")
print("-" * 35)

contacts = []

# Contact 1
print("\nContact 1:")
name1  = input("Name: ")
phone1 = input("Phone: ")
city1  = input("City: ")
contacts.append({"name": name1, "phone": phone1, "city": city1})

# Contact 2
print("\nContact 2:")
name2  = input("Name: ")
phone2 = input("Phone: ")
city2  = input("City: ")
contacts.append({"name": name2, "phone": phone2, "city": city2})

# Contact 3
print("\nContact 3:")
name3  = input("Name: ")
phone3 = input("Phone: ")
city3  = input("City: ")
contacts.append({"name": name3, "phone": phone3, "city": city3})

print("\n✅ Collected", len(contacts), "contacts")
````

````python
contacts
````

````python
print(contacts[0].keys())
print(contacts[1].keys())
print(contacts[2].keys())
````

````python
# Save to CSV
fields = ["name", "phone", "city"]

with open("contacts_input.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields)
    writer.writeheader()
    for c in contacts:
        writer.writerow(c)

print("💾 Saved to contacts.csv")
````

````python
import csv
# Read it back and show the book
print("📒 MY CONTACT BOOK")
print("=" * 40)

with open("contacts_input.csv", "r") as f:
    file_obj = csv.DictReader(f)
    for row in file_obj:
        # print(row["name"], "-", row["phone"], "-", row["city"])
        print(row.get("name"), "-", row.get("phone"), "-", row.get("city"), "-", row.get("Father Name"))

print("=" * 40)
````
